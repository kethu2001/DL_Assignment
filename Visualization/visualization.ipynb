```python
import os
import matplotlib.pyplot as plt

# Path to the train dataset
train_dir = r"C:\Users\Ishan's MSi\Desktop\Dataset\process_data\train"

# Get the class names (folder names)
classes = os.listdir(train_dir)

# Count the number of images in each class
image_counts = {}
for cls in classes:
    image_counts[cls] = len(os.listdir(os.path.join(train_dir, cls)))

# Plot the class distribution
plt.figure(figsize=(12, 8))
plt.bar(image_counts.keys(), image_counts.values(), color='skyblue')
plt.xticks(rotation=90)
plt.title("Class Distribution in the Training Dataset")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import numpy as np

# Function to get image dimensions
def get_image_dimensions(image_path):
    img = Image.open(image_path)
    return img.size

# List to store image dimensions
image_dimensions = []

# Loop through each class folder and image to get dimensions
for cls in classes:
    class_dir = os.path.join(train_dir, cls)
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        image_dimensions.append(get_image_dimensions(img_path))

# Convert dimensions to NumPy array
image_dimensions = np.array(image_dimensions)

# Plot histograms for width and height
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(image_dimensions[:, 0], bins=30, color='blue')
plt.title('Image Width Distribution')
plt.xlabel('Width (pixels)')

plt.subplot(1, 2, 2)
plt.hist(image_dimensions[:, 1], bins=30, color='green')
plt.title('Image Height Distribution')
plt.xlabel('Height (pixels)')
plt.tight_layout()
plt.show()
```

```python
import random
import cv2

# Function to display sample images from different classes
def plot_sample_images(train_dir, classes, num_images=5):
    plt.figure(figsize=(12, 12))
    for i, cls in enumerate(classes[:5]):  # Display first 5 classes
        image_paths = os.listdir(os.path.join(train_dir, cls))
        selected_images = random.sample(image_paths, num_images)
        for j, img_name in enumerate(selected_images):
            img_path = os.path.join(train_dir, cls, img_name)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.subplot(5, num_images, i * num_images + j + 1)
            plt.imshow(img)
            plt.title(cls)
            plt.axis('off')
    plt.show()

plot_sample_images(train_dir, classes)
```
